```python
valores_str = df_homicidios["Hora"].loc[df_homicidios["Hora"].apply(lambda x: isinstance(x, str))]

print(valores_str.unique())
```

se aprecia que hay un sd y los demas son fechas pero en formato cadena. Vamos a convertirlos en datetime.time

```python
df_homicidios["Hora"] = df_homicidios["Hora"].apply(lambda x: np.nan if pd.isnull(x) else x.time() if isinstance(x, datetime) else (x if isinstance(x, time) else (datetime.strptime(x, '%H:%M:%S').time() if isinstance(x, str) and ':' in x else np.nan)))
#Veamos de nuevo frecuencia de valores unicos
fp.unicos_col(df_homicidios["Hora"])
```

Vamos a imputar la media al valor faltante

```python
# Valor específico para llenar los faltantes
valor_especifico = datetime.strptime('1:11:11', '%H:%M:%S').time()

# Rellenar los valores faltantes con el valor específico
df_homicidios["Hora"] = df_homicidios["Hora"].fillna(valor_especifico)
```

```python
from datetime import datetime, timedelta

# Pasar "Hora" a minutos desde la medianoche
df_homicidios["Hora_minutos"] = df_homicidios["Hora"].apply(lambda x: x.hour * 60 + x.minute)

# Calcular la media de los valores existentes en la columna "Hora" en minutos,  
media_minutos = df_homicidios.loc[df_homicidios["Hora"] != valor_especifico, "Hora_minutos"].mean()

# Convertir la media de minutos a formato de tiempo
media_hora = datetime.min + timedelta(minutes=media_minutos)

# Asegurarse de que la media_hora tenga la precisión adecuada (fracciones de segundo)
media_hora = media_hora.replace(microsecond=0)

# Rellenar los valores faltantes con la media recien calculada
df_homicidios["Hora"] = df_homicidios["Hora"].apply(lambda x: media_hora.time() if x == valor_especifico else x)

# Eliminar la columna temporal "Hora_minutos"
df_homicidios = df_homicidios.drop(columns=["Hora_minutos"])
```

# Merge

Vamos a fusionar

```python
df_siniestros = pd.merge(df_homicidios, df_victimas, on="Id", how="left")
df_siniestros
```

Solo queda usar la fecha como fecha de fallecimiento allí donde Fecha_Fallecimiento tiene valores nulos

```python
df_siniestros["Fecha_Fallecimiento"] = df_siniestros.apply(lambda row: row["Fecha"] if pd.isnull(row["Fecha_Fallecimiento"]) else row["Fecha_Fallecimiento"], axis=1)
```

```python
df_siniestros = df_siniestros.rename(columns={"Fecha": "Fecha_Siniestro"})
```

creamos una variable booleana para usarla más adelante, para indicar el semestre

```python
df_siniestros["Semestre"] = df_siniestros["Fecha_Siniestro"].dt.month.apply(lambda month: 1 if month <= 6 else 2)
```

Extracción de dia mes y año en fechas

```python
#Hora
df_siniestros['Hora_Siniestro'] = df_siniestros['Hora'].apply(lambda x: x.hour)

#Dia mes y año de siniestros
df_siniestros['Anio_Siniestro'] = df_siniestros['Fecha_Siniestro'].dt.year
df_siniestros['Mes_Siniestro'] = df_siniestros['Fecha_Siniestro'].dt.month
df_siniestros['Dia_Siniestro'] = df_siniestros['Fecha_Siniestro'].dt.day

#Dia mes y año de fallecimientos
df_siniestros['Anio_Fallecimiento'] = df_siniestros['Fecha_Fallecimiento'].dt.year
df_siniestros['Mes_Fallecimiento'] = df_siniestros['Fecha_Fallecimiento'].dt.month
df_siniestros['Dia_Fallecimiento'] = df_siniestros['Fecha_Fallecimiento'].dt.day
```

obtengamos el dia de la semana en que ocurrieron los eventos

```python
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')
df_siniestros['Dia_Semana_Siniestro'] = df_siniestros['Fecha_Siniestro'].dt.strftime('%A')
df_siniestros['Dia_Semana_Fallecimiento'] = df_siniestros['Fecha_Fallecimiento'].dt.strftime('%A')
```
